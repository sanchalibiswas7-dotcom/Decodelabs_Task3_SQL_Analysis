# Project 3: SQL Data Analysis

```python
import pandas as pd
```

```python
import sqlite3
```

```python
df = pd.read_excel("/content/Cleaned_Ecommerce_Dataset.xlsx")

df.head()
```

```python
conn = sqlite3.connect("ecommerce.db")
```

```python
df.to_sql("orders", conn, if_exists = "replace", index = False)
```

## Displaying the First Five Records

```python
query = """ select * from orders limit 5 """
pd.read_sql(query, conn)
```

## Filtering Data Using the WHERE Clause


```python
query = """ select OrderID, Product, PaymentMethod, TotalPrice from orders where PaymentMethod = 'Credit Card' """
pd.read_sql(query, conn)
```

## Sorting Records Using ORDER BY

```python
query = """ select Product, TotalPrice from orders order by TotalPrice desc """
pd.read_sql(query, conn)
```

## Using Aggregate Functions

```python
query = """ select count(*) as Total_Orders from orders """
pd.read_sql(query, conn)
```

```python
query = """ select sum(TotalPrice) as Total_Revenue from orders """
pd.read_sql(query, conn)
```

```python
query = """ select avg(TotalPrice) as Average_Order_Value from orders """
pd.read_sql(query, conn)
```

## Revenue Analysis by Product Using GROUP BY

```python
query = """ select Product, sum(TotalPrice) as Total_Revenue from orders group by Product """
pd.read_sql(query, conn)
```

## Ranking Products by Revenue

```python
query = """ select Product, sum(TotalPrice) as Total_Revenue from orders group by Product order by Total_Revenue desc """
pd.read_sql(query, conn)
```

## Filtering High-Value Delivered Orders

```python
query = """ select OrderID, Product, TotalPrice, OrderStatus from orders where TotalPrice>1000 and OrderStatus = 'Delivered' """
pd.read_sql(query, conn)
```

## Using HAVING to Filter Grouped Results

```python
query = """ select Product, sum(TotalPrice) as Total_Revenue from orders group by Product having Total_Revenue>50000 """
pd.read_sql(query, conn)
```
